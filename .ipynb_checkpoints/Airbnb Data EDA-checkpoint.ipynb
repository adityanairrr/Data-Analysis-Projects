## Key Observations
**1. We can see a Negative Value in Minimum Nights which is not possible, It has to be removed**\
**2. By looking at the Mean and Max values we can see Outliers in some Columns**

```python
data.isnull().sum()
```

As we Can See there are Lots of Missing Values

```python
data.isnull().sum() / len(data) * 100
```

**License contains 99% Missing Values , So we can remove that column**

```python
data = data.drop(['license'],axis=1)
```

**Since Country Contains Only United States , So Country and Country Code columns are not needed**

```python
data = data.drop(['country'],axis=1)
data = data.drop(['country code'],axis=1)
```

```python
data['last review'] = pd.to_datetime(data['last review'], errors='coerce')
```

**We will Replace the Missing values in reviews per month with 0 and last review with Not Reviewed Because Replacing it with Mode can impact the data**

```python
data['reviews per month'].fillna(0, inplace=True)
data['last review'].fillna('Not Reviewed', inplace=True)
```

```python
data['NAME'].fillna('Not Listed',inplace = True)
data['host name'].fillna('Unknown Host',inplace = True)
data['house_rules'].fillna('Nothing',inplace = True)
```

```python
data['host_identity_verified'].fillna('Yet to be verified', inplace=True)
```

```python
data['cancellation_policy'].fillna('No Policy',inplace=True)
data['instant_bookable'].fillna('False',inplace=True)
```

**We will Fill the missing Values of all these Numerical Columns with Median**

```python
data['availability 365'].fillna(data['availability 365'].median(), inplace=True)
data['calculated host listings count'].fillna(data['calculated host listings count'].median(), inplace=True)
data['review rate number'].fillna(data['review rate number'].median(), inplace=True)
data['number of reviews'].fillna(data['number of reviews'].median(), inplace=True)
data['minimum nights'].fillna(data['minimum nights'].median(), inplace=True)
data['lat'].fillna(data['lat'].median(), inplace=True)
data['long'].fillna(data['long'].median(), inplace=True)
data['id'].fillna(data['id'].median(), inplace=True)
data['host id'].fillna(data['host id'].median(), inplace=True)
data['Construction year'].fillna(data['Construction year'].median(), inplace=True)
```

**Price and Service Fee columns are in Object dtype , So we have to convert it into Float and Then replace Missing Values using Median**

```python
data['price'] = data['price'].str.replace('$', '').str.replace(',', '').astype(float)
data['service fee'] = data['service fee'].str.replace('$', '').str.replace(',', '').astype(float)

# Impute missing values with the median after cleaning
data['price'].fillna(data['price'].median(), inplace=True)
data['service fee'].fillna(data['service fee'].median(), inplace=True)
```

```python
typo_map = {
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan'
}

# Use the .replace() method to apply the mapping to the column
data['neighbourhood group'] = data['neighbourhood group'].replace(typo_map)
```

```python
# Maximum Nights contains some negative values which is not possible so to remove that We use this code
initial_rows = len(data)
data = data[data['minimum nights'] > 0].copy()
```

```python
# No of rows Removed because they were less than 0 
initial_rows - len(data)
```

**These Four columns (calculated host listings count,minimum nights,number of reviews,reviews per month) Have Large Outliers So to Remove them We Use Capping**

```python
capping_cols = ['calculated host listings count', 'minimum nights', 'number of reviews', 'reviews per month']
cap_percentile = 0.995 # Using the 99.5th percentile

# Calculate the capping thresholds (Q99.5) for the selected columns
caps = data[capping_cols].quantile(cap_percentile)

# Apply the capping (replace values greater than the cap with the cap value)
for col in capping_cols:
    # Use np.where for conditional replacement
    data[col] = np.where(data[col] > caps[col], caps[col], data[col])

# Value Greater than Upper limit then Replace it with Upper Limit else Leave it as it is
```

```python
# --- Verification (Display new statistics) ---
print("\nDescriptive Statistics for Columns After Capping:")
print(data[capping_cols].describe().to_markdown(numalign="left", stralign="left"))
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(data['reviews per month'])
plt.show()
```

**Even after doing Capping We can see Lots of Outliers , So now we will have to Apply Log Transformation**

```python
log_cols = [
    'minimum nights', 
    'number of reviews', 
    'reviews per month',
    'calculated host listings count'
]

# Apply Log Transformation (log1p = log(x+1))
print("Applying log(x+1) transformation...")

for col in log_cols:
    new_col_name = f'log_{col}'
    
    # Check if the column exists before transforming (safety check)
    if col in data.columns:
        # np.log1p(x) computes log(1+x), which is stable for small x and handles min value 0 gracefully
        data[new_col_name] = np.log1p(data[col])
        print(f"Created new column: {new_col_name}")
    else:
        print(f"Warning: Column '{col}' not found in DataFrame.")
```

```python
data['log_price'] = np.log1p(data['price'])
```
