```python
#Avg Scores in diff Subjects by gender 
gender_performance = data.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()
gender_performance2 = gender_performance.melt(id_vars='gender', var_name='Subject', value_name='Mean Score')

plt.figure(figsize=(10, 6))
sns.barplot(data=gender_performance2, x='Subject', y='Mean Score', hue='gender')

plt.legend(
    title='Gender', 
    bbox_to_anchor=(1.05, 1), # Places the legend to the right of the plot
    loc='upper left', 
    borderaxespad=0.
)
```

```python
#Avg Scores in diff Subjects by Ethnicity
race_performance = data.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean().reset_index()
race_performance1 = race_performance.melt(id_vars='race/ethnicity', var_name='Subject', value_name='Mean Score')
plt.figure(figsize=(10, 6))
sns.barplot(data=race_performance1 ,x='Subject', y='Mean Score', hue='race/ethnicity')

plt.legend(
    title='race/ethnicity', 
    bbox_to_anchor=(1.05, 1), # Places the legend to the right of the plot
    loc='upper left', 
    borderaxespad=0.
)
```

```python
#Checking impact of test prep course on avg scores
data['average score'] = data[['math score', 'reading score', 'writing score']].mean(axis=1)

completed = data[data['test preparation course'] == 'completed']['average score']
none = data[data['test preparation course'] == 'none']['average score']

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='test preparation course', y='average score', palette='Set2')
plt.title('Impact of Test Prep Course on Average Score')
plt.show()
```

```python
#Avg Score by Education Level and Race
pivot_table = data.pivot_table(values='average score', index='parental level of education', columns='race/ethnicity', aggfunc='mean')
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt=".1f")
plt.title('Average Score by Education Level and Race')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = data[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Math, Reading, and Writing Scores')
```

```python
#Checking impact of lunch on Avg scores
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='lunch', y='average score', palette='Set2')
plt.title('Impact of Lunch on Average Score')
plt.show()
```

# Key Observations

**1. General Performance**

* The overall mean scores are roughly 66 for Math, 69 for Reading, and 68 for Writing.
* There is a very high correlation (0.95) between Reading and Writing. If a student is good at reading, they are almost certainly good at writing.

**2. Gender Based**

* There are 48% Male and 51% Female
* Males perform better in Math as there avg scores are better than females
* Females perform better in Reading and Writing

**3. Parental Education Influence**

* There is a clear upward trend: as the level of parental education increases (from High School -> Associate -> Bachelor -> Master), the student's average score typically increases.

**4. Race/Ethnicity Wise**

* Many students are from the Group C (almost 31%) and Group D (almost 26%) ethnicity
* Students in Group E consistently showed the highest average scores
* Students in Group A generally had the lowest average scores across all three subjects.

**5. Other Obs**

* Students who completed the test preparation course scored significantly higher in all subjects
* Students receiving "standard" lunch consistently outperformed those with "free/reduced" lunch.
