**Now, Lets do Analysis of a Particular match**

```python
match_id = 734049 
match_df = ipl[ipl['match_id'] == match_id].copy().sort_values(['inning', 'over', 'ball'])
```

**Selected the Final of 2014 IPL - Kolkata Knight Riders vs Kings XI Punjab**

```python
match_df
```

```python
# Generating the 'Live Score' after every ball
match_df['current_score'] = match_df.groupby('inning')['total_runs'].cumsum()
match_df['wickets_lost'] = match_df.groupby('inning')['is_wicket'].cumsum()
```

```python
# Logic: Target is Inning 1 Total + 1. RRR is only for Inning 2.
inning1_total = match_df[match_df['inning'] == 1]['total_runs'].sum()
target = inning1_total + 1
```

```python
inning1_total
```

```python
target
```

```python
#Calculating Required Run Rate
def get_rrr(row):
    if row['inning'] == 2:
        balls_bowled = (row['over'] - 1) * 6 + row['ball']
        balls_left = 120 - balls_bowled
        runs_needed = target - row['current_score']
        return (runs_needed / balls_left) * 6 if balls_left > 0 else 0
    return 0

match_df['RRR'] = match_df.apply(get_rrr, axis=1)
```

```python
## Pressure index i.e Higher wickets = higher pressure even if RRR is low.
match_df['pressure_index'] = match_df['RRR'] * (match_df['wickets_lost'] + 1)
```

```python
match_df
```

```python
# Final summary
over_summary = match_df.groupby(['inning', 'over']).agg({
    'total_runs': 'sum',           # Runs scored in this over
    'current_score': 'last',       # Score at the end of this over
    'wickets_lost': 'last',        # Wickets at the end of this over
    'RRR': 'last',                 # rrr during this over
    'pressure_index': 'max'        # Peak pressure during the over
}).reset_index()
```

```python
over_summary
```

```python
match_df['phase'] = pd.cut(match_df['over'], bins=[0, 6, 15, 20], labels=['Powerplay', 'Middle', 'Death'])
phase_stats = match_df.groupby(['inning', 'phase'])['total_runs'].sum()
```

**Lets Look at some Overall Stats**

```python
#Mapping teams with double entries name 
team_fix = {
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Royal Challengers Bengaluru' : 'Royal Challengers Bangalore'
    
    # Optional: depends if you want to group them
}

# Apply the fix to all relevant columns at once
cols_to_fix = ['team1', 'team2', 'toss_winner', 'winner']
for col in cols_to_fix:
    matches[col] = matches[col].replace(team_fix)
```

```python
# Calculate % of matches won by the toss winner
toss_win_match_win = matches[matches['toss_winner'] == matches['winner']]
toss_factor_pct = (len(toss_win_match_win) / len(matches)) * 100

print(f"Overall IPL Toss Factor: {toss_factor_pct:.2f}%")
```

```python
# Define the Home Grounds
home_ground_map = {
    'Mumbai Indians': 'Wankhede Stadium',
    'Chennai Super Kings': 'MA Chidambaram Stadium, Chepauk',
    'Royal Challengers Bengaluru': 'M. Chinnaswamy Stadium',
    'Kolkata Knight Riders': 'Eden Gardens',
    'Rajasthan Royals': 'Sawai Mansingh Stadium',
    'Delhi Capitals': 'Arun Jaitley Stadium',
    'Punjab Kings': 'Punjab Cricket Association IS Bindra Stadium, Mohali',
    'Sunrisers Hyderabad': 'Rajiv Gandhi International Stadium, Uppal',
    'Rising Pune Supergiant': 'Maharashtra Cricket Association Stadium',
    'Gujarat Titans': 'Narendra Modi Stadium, Ahmedabad',
    'Lucknow Super Giants': 'Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium, Lucknow',
    'Kochi Tuskers Stadium' : 'Nehru Stadium',
    'Gujarat Lions' : 'Saurashtra Cricket Association Stadium',
    
        
}

# Identify Home Matches vs Away Matches
def get_match_type(row):
    # Check if Team 1 or Team 2 is playing at their mapped home ground
    if row['venue'] == home_ground_map.get(row['team1']):
        return row['team1']
    if row['venue'] == home_ground_map.get(row['team2']):
        return row['team2']
    return 'Neutral'

matches['home_team'] = matches.apply(get_match_type, axis=1)

# Calculate Win % at Home
home_matches = matches[matches['home_team'] != 'Neutral']
home_stats = home_matches.groupby('home_team').agg(
    total_home_games=('id', 'count'),
    home_wins=('winner', lambda x: (x == home_matches.loc[x.index, 'home_team']).sum())
)

home_stats['home_win_rate'] = (home_stats['home_wins'] / home_stats['total_home_games']) * 100
print(home_stats.sort_values(by='home_win_rate', ascending=False))
```

```python
# Venue-wise Average Scores and Winning Trends
venue_stats = matches.groupby('venue').agg({
    'id': 'count',                 # Total Matches
    'toss_decision': lambda x: x.value_counts().index[0], # Most common decision
    'winner': 'count'
}).rename(columns={'id': 'Total Matches'})

# To get Avg Score, we need the deliveries dataset
avg_score_venue = ipl.groupby(['venue', 'match_id', 'inning'])['total_runs'].sum().reset_index()
venue_par_scores = avg_score_venue[avg_score_venue['inning'] == 1].groupby('venue')['total_runs'].mean()

print("Venue Par Scores (1st Innings Average):")
print(venue_par_scores.sort_values(ascending=False))
```

```python
deliveries['current_score'] = deliveries.groupby(['match_id', 'inning'])['total_runs'].cumsum()
deliveries['wickets_lost'] = deliveries.groupby(['match_id', 'inning'])['is_wicket'].cumsum()

#Target calculate
match_targets = deliveries[deliveries['inning'] == 1].groupby('match_id')['total_runs'].sum().reset_index()
match_targets.columns = ['match_id', 'target']

deliveries = pd.merge(deliveries, match_targets, on='match_id', how='left')

#RRR 
def calculate_rrr_bulk(row):
    if row['inning'] == 2:
        balls_bowled = (row['over'] - 1) * 6 + row['ball']
        balls_left = 120 - balls_bowled
        runs_needed = row['target'] - row['current_score']
        if balls_left > 0:
            return round((runs_needed / balls_left) * 6, 2)
    return 0

deliveries['RRR'] = deliveries.apply(calculate_rrr_bulk, axis=1)

#Exporting
deliveries.to_csv('ball_by_ball_enriched.csv', index=False)
```
